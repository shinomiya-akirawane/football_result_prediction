```python
import pandas as pd

season_data_path = 'season_data.csv'
team_abilities_path = 'team_abilities.csv'

season_data = pd.read_csv(season_data_path)
team_abilities = pd.read_csv(team_abilities_path)

seasons = sorted(season_data['Season'].unique(), reverse=True)
clubs = season_data['Club'].unique()

historical_abilities = {club: {} for club in clubs}

# 当前赛季能力值
current_season = '2023-2024'
current_abilities = team_abilities.set_index('Club')

# 获取最低能力值
lowest_abilities = team_abilities.nsmallest(1, ['GK_Ability', 'team_offensive_ability', 'team_possession_ability', 'team_defensive_ability']).iloc[0].to_dict()

# 迭代计算每个赛季的能力值
for i, season in enumerate(seasons):
    if season == current_season:
        continue  # 跳过当前赛季

    season_stats = season_data[season_data['Season'] == season]
    relegated_teams = season_stats[season_stats['Rk'] > 17]['Club']

    # 针对每个俱乐部计算能力值
    for club in clubs:
        # 获取球队过去赛季的数据
        past_season_stats = season_data[(season_data['Season'] == season) & (season_data['Club'] == club)]

        if not past_season_stats.empty:
            # 计算过去赛季的能力值因子
            factors = {
                'GK_Ability': past_season_stats['GA/MP'].iloc[0] / (current_abilities.loc[club, 'GA/MP'] if club in current_abilities.index else 1),
                'team_offensive_ability': past_season_stats['GF/MP'].iloc[0] / (current_abilities.loc[club, 'GF/MP'] if club in current_abilities.index else 1),
                'team_possession_ability': past_season_stats['Pts/MP'].iloc[0] / (current_abilities.loc[club, 'Pts/MP'] if club in current_abilities.index else 1),
                'team_defensive_ability': past_season_stats['GA/MP'].iloc[0] / (current_abilities.loc[club, 'GA/MP'] if club in current_abilities.index else 1),
            }

            # 应用因子计算能力值
            historical_abilities[club][season] = {
                'GK_Ability': (current_abilities.loc[club, 'GK_Ability'] if club in current_abilities.index else lowest_abilities['GK_Ability']) * factors['GK_Ability'],
                'team_offensive_ability': (current_abilities.loc[club, 'team_offensive_ability'] if club in current_abilities.index else lowest_abilities['team_offensive_ability']) * factors['team_offensive_ability'],
                'team_possession_ability': (current_abilities.loc[club, 'team_possession_ability'] if club in current_abilities.index else lowest_abilities['team_possession_ability']) * factors['team_possession_ability'],
                'team_defensive_ability': (current_abilities.loc[club, 'team_defensive_ability'] if club in current_abilities.index else lowest_abilities['team_defensive_ability']) * factors['team_defensive_ability'],
            }
        else:
            # 为降级队伍设置初始能力值
            if club in relegated_teams.values:
                rank = list(relegated_teams).index(club) + 1
                penalty = 0.9 ** rank
                historical_abilities[club][season] = {k: v * penalty for k, v in lowest_abilities.items()}

all_historical_abilities = pd.DataFrame()
for club, abilities in historical_abilities.items():
    df = pd.DataFrame.from_dict(abilities, orient='index')
    df['Club'] = club
    all_historical_abilities = pd.concat([all_historical_abilities, df])

all_historical_abilities.reset_index(inplace=True)
all_historical_abilities.rename(columns={'index': 'Season'}, inplace=True)
all_historical_abilities.set_index(['Club', 'Season'], inplace=True)

all_historical_abilities.to_csv('all_historical_abilities.csv')
```
